## Conditional execution

- Execute code depending on whether some condition evaluates to `True`
- Syntax: `if`/`elif`/`else` statement
- Statement body must be indented

*Example*:

- Compare arithmetic expressions for equality (`==`)

````python
if 2*2 == 3:
    print('Python knows arithmetic!')
elif 2*2 == 4:
    print('Elif statement evaluated')
else:
    print('Condition is not true')

# The first True statement is printed
````

### Truth value testing

| Expression    | Description |
| ------------- | ----------- |
| `==`          | Equal. Works for numerical values, strings, etc. |
| `!=`          | Not equal. Works for numerical values, strings, etc. |
| `>`, `>=`,`<=`, `<` | Usual comparison of numerical values |
|  `a is b`, `a is not b`   | Test identity. `a is b` is `True` if `a` and `b` are the same object |
| `a in b`, `a not in b`    | Test whether `a` is or is not included in collection `b` |
| `if obj`, `if not obj`    | Most Python objects evaluate to `True` or `False` in an intuitive fashion (see below) |

Additionally, logical operators allow us to combine two logical values:

| Expression | Description |
| ---------- | ----------- |
| `a and b`  | `True` if both `a` and `b` are `True` |
| `a or b`   | `True` if at least one of `a` or `b` is `True` |

*Examples*:

- Illustrate `is`, `==`, and `in` for lists
- Illustrate `in` using dictionary keys

````python
# Define a list
a = [1, 2]
b = a
````

````python
a is b
````

````python
b = a.copy()
a is b  # b is not a, b is a COPY of a
````

````python
# Elementwise comparison
a == b  # Elements of a and b are the same
````

````python
# Check whether something is in a container
1 in a
````

````python
3 not in a
````

````python
not 3 in a
````

````python
if 1 in a:
    print('List a contains the element 1')
````

````python
# Create a variable that is an empty string
var = ""

# Test whether the string is empty
if var == "":
    print('Empty string encountered')
````

***
Most objects evaluate to `True` or `False`
in an `if` statement:
```python
if obj:
    # do something if obj evaluates to True
```
The rules are quite intuitive (also see workshop 1): an object evaluates to `False` if

-   it has a numerical type and is `0` (or `0.0`, or complex `0+0j`)
-   it is an empty string `''`
-   it is an empty collection (tuple, list, dictionary, etc.)
-   it is of logical (boolean) type and has value `False`
-   it is `None`, a special built-in value used to denote that
    a variable does not reference anything.

*Examples*:

- Check for zero values
- Check for empty strings
- Check for empty collections (strings, dictionaries)
- Check boolean flags

````python
# Empty string
var = ""
if not var:
    print('empty string')

var = "abc"
if var:
    print('Not an empty string')
````

````python
# Numbers evaluate to True if they are not 0 or 0.0
var = 10
if var:
    print('Number is non-zero')

var = 0
if not var:
    print('Number is zero')
````

````python
a = [1]
if a:
    print('Collect is not empty')
````

````python
d = {'key1': 'value1'}
if d:
    print('Collection is not empty')

# Unpythonic way, because it is more syntax than you need
if len(d) > 0:
    print('Collection is not empty')
````

**Important exception:** NumPy arrays 
-   Cannot be used in `if` statements directly because the truth value of an array with more than one element is ambiguous.

*Examples:*

- Check expressions involving NumPy arrays using `np.all()` and `np.any()`

````python
import numpy as np

# Create a numpy array of 5 zeros
arr = np.zeros(5)
````

````python
if len(arr) > 0:
    print('Numpy array has more than zero elements')
````

````python
if np.all(arr == 0):
    print('All elements are zero')

if np.any(arr != 0):
    print('At least one element is not zero')
````

### Conditional expressions

- Compact expressions that can be used for conditional assignment
- Syntax: 
    ```python
    <value-if-true> if <condition> else <value-if-false>
    ```
- Even shorter: 
    ```python
    <value> or <default>
    ```

````python
var = 1

# Check whether value in var is odd or even
if (var % 2) == 0:
    message = 'Var is even'
else:
    message = 'Var is odd'
print(message)
````

````python
# Use conditional expression to write this as a one-liner
var = 2
message = 'Var is even' if (var % 2) == 0 else 'Var is odd'
print(message)
````

## Loops

### The `for` loop

- Iterates over arbitrary collections of items
- Special case: iterate over integers using [`range()`](https://docs.python.org/3/library/functions.html#func-range)

*Examples:*

- Iterate over range of integers
- Iterate over tuple `cities = ('Bergen', 'Oslo', 'Stavanger')`

````python
# Iterate over the integers 0, ..., 4
for i in range(5):
    print(i)
````

````python
for i in range(1, 5, 2):
    print(i)
````

````python
for i in range(4, 0, -1):
    print(i)
````

````python
cities = ('Bergen', 'Oslo', 'Stavanger')

# Iterate over integers (some old languages can ONLY iterate over integers)
for i in range(len(cities)):
    print(cities[i])
````

````python
# In python: iterate over cities directly
for city in cities:
    print(city)
````

#### Looping over dictionaries

- Default: loop over keys
- Can loop over values using the `.values()`
- Most common usage: loop over keys + values using `.items()`

*Example:*

- Loop over dictionary `dct = {'key1': 'value1', 'key2': 'value2'}`

````python
dct = {'key1': 'value1', 'key2': 'value2'}
# By default looping is over the keys
for key in dct:
    print(key)
````

````python
# Loop over values
for val in dct.values():
    print(val)
````

````python
# Loop over keys and values
for key in dct:
    value = dct[key]
    print(f'{key}: {value}')
````

````python
# Python special syntax
for key, value in dct.items():
    print(f'{key}: {value}')
````

### The `while` loop

- Less commonly used than `for` loop
- Used when number of iterations is unknown ex ante
- Used when iteration is *not* over some sequence of items 

````python
z = 1.001
i = 0

while z < 100.0:
    z = z*z + 0.234
    # Increment iteration counter
    i += 1

print(f'Number of iterations: {i}')
````

````python
while True:
    break
````

### Advanced looping

- Use `enumerate()` to generate an iteration index
- Use `continue` to skip the remainder of an iteration
- Use `break` to terminate the loop
- Use `zip()` to iterate over more than one collection in parallel

*Examples:*

- Illustrate `zip()` with lists
    ```python
    capitals = ['Bergen', 'Stavanger', 'Trondheim']
    regions = ['Vestland', 'Rogaland', 'TrÃ¸ndelag']
    ```

````python
cities
````

````python
# Iterated over cities
for city in cities:
    print(city)
````

````python
for i in range(len(cities)):
    print(f'City {i} is {cities[i]}')
````

````python
# Pythonic version of the previous loop
for i, city in enumerate(cities):
    print(f'City {i} is {city}')
````

````python
for city in cities:
    if city == 'Oslo':
        # Skip current iteration
        continue
    print(city)
````

````python
for city in cities:
    if city == 'Bergen':
        print('Found Bergen')
        # Terminate loop
        break
````

````python
capitals = ['Bergen', 'Stavanger', 'Trondheim']
regions = ['Vestland', 'Rogaland', 'TrÃ¸ndelag']
````

````python
# Iterate over regions and capitals at once
for capital, region in zip(capitals, regions):
    print(f'The capital of {region} is {capital}')
````

## List comprehensions

- Compact expression used to create lists, tuples, dictionaries
- Packs the functionality of a `for` loop into a single line 
- Can be combined with conditions using `if`
- List comprehensions can be nested, just like loops

*Examples:*

- List of squares of 0,...,4 created with list comprehensions vs loops
- Dictionary mapping numbers to squares 

````python
# Initialize an empty list
items = []
for i in range(5):
    # Append square to list
    items.append(i**2)

# Print list of squares
print(items)
````

````python
# List comprehension
items = [i**2 for i in range(5)]
items
````
